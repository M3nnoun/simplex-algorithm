```python
## fiding the input
def get_the_form():
    print("<<Please Enter the form canonice>>")
    contraints_info=input("please enter the numbers of varaibles, and the number of contraints").split()
    contraints_info=list(map(int,contraints_info))
    print("the form must be in ax+by+cz ....<=B [last number is the element in B matrixe]")
    counter=0
    lines=[]
    while(counter<contraints_info[1]):
        inquation=input(f"please enter the inquation numbe [{counter+1}]").split()
        inquation=list(map(int,inquation))
        lines.append(inquation)
        counter+=1
    #let get the objectives function
    cofetions=input('enter the cofietion of the objective function in the same order').split()
    cofetions=list(map(int,cofetions))
    return lines,cofetions
```

```python
def proplem_builder():
    lines,cofetions=get_the_form()
    number_equation=len(lines)
    ## add ecart
    matrix_contraint=np.array(lines)
    matrixe_b=matrix_contraint[:,-1]
    proplem=np.hstack((matrix_contraint[:,:-1],np.eye(number_equation),matrixe_b[:,np.newaxis]))
    ##create the last line:
    line_objective=cofetions+[0 for _ in range(number_equation+1)]
    return np.vstack((proplem,line_objective))    
```
